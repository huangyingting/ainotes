```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
from gptpdf import parse_pdf
import os

def list_pdf_files(directory):
    """
    List all PDF files under the given directory, including files in subdirectories.
    """
    pdf_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.pdf'):
                pdf_files.append(os.path.join(root, file))
    return pdf_files

model = 'azure_gpt-4o'
dir = 'pdfs'
pdf_files = list_pdf_files(dir)

for pdf_file in pdf_files:
    print(f'Processing {pdf_file}')
    base_name, _ = os.path.splitext(os.path.basename(pdf_file))  # Split the extension
    output_dir = os.path.join('output', base_name)
    content, image_paths = parse_pdf(pdf_file, 
                                 output_dir=output_dir, 
                                 api_key=os.getenv("AZURE_OPENAI_API_KEY"), 
                                 base_url=os.getenv("AZURE_OPENAI_ENDPOINT"), 
                                 model=model)
    os.rename(os.path.join(output_dir, 'output.md'),
              os.path.join(output_dir, base_name + '.md'),)
```
